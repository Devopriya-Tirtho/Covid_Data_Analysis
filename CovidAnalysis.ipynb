```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv('owid-covid-data.csv',error_bad_lines=False)
```

```python
dataset.columns
```

```python
pd.unique(dataset['location'])
```

```python
start_date = '2021-06-30'
end_date = '2021-07-30'
```

```python
dataset.head()
```

```python
mask = (dataset['date'] > start_date) & (dataset['date'] <= end_date)
```

```python
mask
```

```python
df = dataset.loc[mask]
df
```

```python
main_df=df[['location','date','new_cases','new_deaths','new_cases_per_million','new_deaths_per_million','people_fully_vaccinated_per_hundred']]
```

```python
main_df.head()
```

```python
main_df.dropna()
```

```python
main_df = main_df.set_index("location")
main_df.head()
```

```python
main_df
```

```python
main_df = main_df.drop(['Afghanistan', 'Africa', 'Albania', 'Algeria', 'Andorra', 'Angola', 'Anguilla', 'Antigua and Barbuda'])
```

```python
main_df
```

```python
main_df1=main_df.loc([['United States','India','Brazil','Russia','France','United Kingdom','Turkey','Argentina','Colombia','Spain','Italy','Iran','Germany','Mexico','Indonesia','Poland','Canada','Bangladesh'] , : ])
```
